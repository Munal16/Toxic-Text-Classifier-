```python
import pandas as pd

df = pd.read_json(
    "hf://datasets/mteb/toxic_conversations_50k/train.jsonl",
    lines=True
)
```

```python
df
```

```python
import matplotlib.pyplot as plt

class_counts = df['label'].value_counts()

plt.figure()
class_counts.plot(kind='bar')
plt.xlabel('Class Label')
plt.ylabel('Number of Samples')
plt.title('Class Distribution of Toxic vs Non-Toxic Samples')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Non-Toxic', 'Toxic']
sizes = df['label'].value_counts(normalize=True) * 100

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage Distribution of Toxic vs Non-Toxic Samples')
plt.show()
```
